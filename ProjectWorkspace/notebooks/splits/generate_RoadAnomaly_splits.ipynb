```python
#===============================================================
# DEBUG TOGGLE SWITCH
def d(*args, **kwargs):
    if d.ON:
        print(*args, **kwargs)
d.ON = True
#===============================================================


#===============================================================00
import os
from tqdm import tqdm
#===============================================================0


#===============================================================
# PATHS SETTING
root_dir = "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/Datasets"
frames_dir = os.path.join(root_dir, "RoadAnomaly_jpg/frames")
output_txt_path = os.path.join(root_dir, "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/ProjectWorkspace/splits/roadAnomaly_valid_pairs.txt")
#============================================================================================================
# BUILDING THE PAIRS
valid_pairs = []

for fname in tqdm(sorted(os.listdir(frames_dir))):
  if fname.lower().endswith(".jpg"):
    name_wo_ext = os.path.splitext(fname)[0]
    label_dir = os.path.join(frames_dir, name_wo_ext + ".labels")
    label_path = os.path.join(label_dir, "labels_semantic.png")

    if os.path.exists(label_path):
      rgb_path = os.path.join(frames_dir, fname)
      valid_pairs.append(rgb_path)
    else:
      d(f"[SKIP] No label found for: {fname}")

d(f"Found {len(valid_pairs)} valid pairs")
#================================================================================================


#=========================================================================0
# BUILDING .txt FILE
with open(output_txt_path, "w") as f:
  for path in valid_pairs:
    f.write(path + "\n")

d(f"Saved split file to {output_txt_path}")



```
