```python
#===============================================================
# DEBUG TOGGLE SWITCH

def d(*args, **kwargs):
    if d.ON:
        print(*args, **kwargs)

d.ON = True  # <- Set False to disable debug
#===============================================================
if __name__ == "__main__":
  try:
    import tqdm
  except ImportError:
    !pip install tqdm
    import tqdm

#================================================================
import os
#===============================================================




#================================================================00
def generate_laf_split(images_root, output_txt_path):
  count = 0
  all_valid = []

  for city in sorted(os.listdir(images_root)):
    city_path = os.path.join(images_root, city)
    if not os.path.isdir(city_path):
      continue

    for file in os.listdir(city_path):
      if file.endswith("_leftImg8bit.png"):
        base_name = file.replace("_leftImg8bit.png", "")
        all_valid.append(f"{city}/{base_name}")
        count += 1

    with open(output_txt_path, "w") as f:
      for item in sorted(all_valid):
        f.write(f"{item}\n")

    d(f"[INFO] Found {count} valid LostAndFound test scenes")
    d(f"[INFO] Split file saved to: {output_txt_path}")
#=====================================================================

#=================================================================
generate_laf_split(
    images_root = "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/Datasets/LostAndFound/leftImg8bit/test",
    output_txt_path = "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/ProjectWorkspace/splits/lostAndFound_splits.txt"
)
```
