```python
#===============================================================
# DEBUG TOGGLE SWITCH
def d(*args, **kwargs):
    if d.ON:
        print(*args, **kwargs)
d.ON = False  # ‚Üê set True to enable debug output
#=============================================================

import os
import random

#================================================================
# CONFIGURATION
ROOT_GT_FINE = "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/Datasets/leftImg8bit_trainvaltest/gtFine_trainvaltest/gtFine"
SPLIT_DIR = "/content/drive/MyDrive/MAGISTRALE/ANNO 1/Computer Vision/Project/RoadObstacleDetection/ProjectWorkspace/splits"
os.makedirs(SPLIT_DIR, exist_ok=True)
d("Created or found split directory:", SPLIT_DIR)

# split for training
train_dir = os.path.join(ROOT_GT_FINE, "train")
train_list = []

for city in os.listdir(train_dir):
  city_path = os.path.join(train_dir, city)
  for file in os.listdir(city_path):
    if file.endswith("_gtFine_labelIds.png"):
      base = file.replace("_gtFine_labelIds.png", "")
      rel_path = f"{city}/{base}"
      train_list.append(rel_path)

d("Found ", len(train_list), "train samples")
with open(os.path.join(SPLIT_DIR, "train_list.txt"), "w") as f:
  f.write("\n".join(sorted(train_list)))  # saving the sorted list of training samples to this text file
d("Saved train_list.txt file")


# split for validation and calibration
val_dir = os.path.join(ROOT_GT_FINE, "val")
val_list = []

for city in os.listdir(val_dir):
  city_path = os.path.join(val_dir, city)
  for file in os.listdir(city_path):
    if file.endswith("_gtFine_labelIds.png"):
      base = file.replace("_gtFine_labelIds.png", "")
      rel_path = f"{city}/{base}"
      val_list.append(rel_path)

d("Found ", len(val_list), "validation samples")
random.seed(42)  # fixed for absolute reproducibility, randomic behaviour becomes deterministic, because the samples should not change across different runs
random.shuffle(val_list) # now they get mixed

val_early = val_list[:400] #first 400 for validation (80%)
calib_cp = val_list[400:] #last 100 for calibration (20%)

with open(os.path.join(SPLIT_DIR, "val_early.txt"), "w") as f:
  f.write("\n".join(sorted(val_early))) # saving the sorted list of validation samples to this text file
d("Saved val_early.txt file with", len(val_early), "samples")

with open(os.path.join(SPLIT_DIR, "calib_cp.txt"), "w") as f:
  f.write("\n".join(sorted(calib_cp))) # saving the sorted list of calibration samples to this text file
d("Saved calib_cp.txt file with", len(calib_cp), "samples")
```
